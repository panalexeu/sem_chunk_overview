#### edu mode

```python
import tiktoken
from tiktoken._educational import * 

text = '''A Student's Guide to Startups

Want to start a startup? Get funded by Y Combinator.

October 2006

This essay is derived from a talk at MIT.

Till recently graduating seniors had two choices: get a job or go to grad school. I think there will increasingly be a third option: to start your own startup. But how common will that be?
'''

enc = SimpleBytePairEncoding.from_tiktoken("cl100k_base")
enc.encode(text)
```

#### comparing tokens amount to characters, and words

```python
enc = tiktoken.get_encoding('cl100k_base')
tokens = enc.encode(text)
print(f'chars: {len(text)} | tokens: {len(tokens)} | words: {len(text.split())} ')
tokens
```

```python
enc.decode(tokens)
```

```python
stmt = len(tokens) == len(set(tokens))
print(len(tokens), len(set(tokens)))
```

```python
dict_ = {}  
for token in tokens:
    if token not in dict_:
        count = tokens.count(token)
        if count > 1:  
            dict_[token] = count

dict_
```

```python
dec_dict = {}
for key, val in dict_.items():
   dec_dict[key] = enc.decode([key])

dec_dict
```

### sentence transformer tokenizer

```python
from sentence_transformers import SentenceTransformer

model = SentenceTransformer('sentence-transformers/all-MiniLM-L6-v2')

model.tokenizer
```

```python
model.tokenizer.tokenize(text)
```

```python
str_tokens = model.tokenizer.tokenize(text)
tokens = model.tokenize(text)

print(len(str_tokens), len(tokens))
tokens
```

```python
with open('./paul_graham_essay.txt', 'r') as file:
    text = file.read()

text
```

```python
tokens = model.tokenizer.tokenize(text) 

for i in range(0, len(tokens), 16):
    print(' '.join(tokens[i:i+16]))
    print('=' * 50)
```

```python
tokens = enc.encode(text)
step = 50

for i in range(0, len(tokens), step):
    print(enc.decode(tokens[i:i+step]))
    print('=' * 50)
```
