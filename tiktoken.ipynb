#### edu mode

```python
import tiktoken
from tiktoken._educational import * 

text = '''A Student's Guide to Startups

Want to start a startup? Get funded by Y Combinator.

October 2006

This essay is derived from a talk at MIT.

Till recently graduating seniors had two choices: get a job or go to grad school. I think there will increasingly be a third option: to start your own startup. But how common will that be?
'''

enc = SimpleBytePairEncoding.from_tiktoken("cl100k_base")
enc.encode(text)
```

#### comparing tokens amount to characters, and words

```python
enc = tiktoken.get_encoding('cl100k_base')
tokens = enc.encode(text)
print(f'chars: {len(text)} | tokens: {len(tokens)} | words: {len(text.split())} ')
tokens
```

```python
enc.decode(tokens)
```

```python
stmt = len(tokens) == len(set(tokens))
print(len(tokens), len(set(tokens)))
```

```python
dict_ = {}  
for token in tokens:
    if token not in dict_:
        count = tokens.count(token)
        if count > 1:  
            dict_[token] = count

dict_
```

```python
dec_dict = {}
for key, val in dict_.items():
   dec_dict[key] = enc.decode([key])

dec_dict
```
